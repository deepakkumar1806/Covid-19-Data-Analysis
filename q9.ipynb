```python
import pandas as pd
import numpy as np
import json
from datetime import datetime
from datetime import timedelta
```

```python
#list of common districts
l = []
for x in neighbor_districts:
    if x[1].lower() in name:
        l.append(x)
neighbor_districts = l
```

```python
#data from census and vaccinaiton file
total_population = []
vaccination_on_13 = []
vaccination_on_14 = []
for district in neighbor_districts:
    k = list(vaccination_data["District"]).index(district[1])
    vaccination_on_13.append(int(vaccination_data["13/08/2021.3"][k]))
    vaccination_on_14.append(int(vaccination_data["14/08/2021.3"][k]))
    k = list(census_data["Name"]).index(district[1].lower())
    total_population.append(int(census_data["TOT_P"][k]))
```

```python
#population left and rate of vaccination for each district
population_left = []
rate_of_vaccination = []
for i in range(len(total_population)):
    pop_left = total_population[i]-vaccination_on_14[i]
    if pop_left < 0:
        population_left.append(0)
    else:
        population_left.append(pop_left)
    rate_of_vaccination.append(vaccination_on_14[i]-vaccination_on_13[i])
```

```python
#list of districts according to district ids
districtlist = []
for i in range(len(neighbor_districts)):
    districtlist.append(neighbor_districts[i][0])
```

```python
#resulted dataframe creation
df = {
    "districtid" : districtlist,
    "populationleft" : population_left,
    "rateofvaccination" : rate_of_vaccination
}

df = pd.DataFrame(df)
#adding column for state code
state_code = []
for district in df["districtid"]:
    state_code.append(district.split("_")[0]) 
    
df["stateid"] = state_code
```

```python
df
```

```python
#list of states
state = []
for i in neighbor_districts:
    state.append(i[0].split("_")[0])
state = list(np.unique(state))
```

```python
#state wise information
population_left_state = []
rate_of_vaccination_state = []
for s in state:
    x = df.query("stateid == @s")
    population_left_state.append(np.sum(x["populationleft"]))
    rate_of_vaccination_state.append(np.sum(x["rateofvaccination"]))
```

```python
#dates of vaccination completion
days = []
for i in range(len(state)):
    days.append(np.ceil(population_left_state[i]/rate_of_vaccination_state[i]))
    
dates = []
for i in range(len(days)):
    Begindatestring = "2021-08-14"
    Begindate = datetime.strptime(Begindatestring, "%Y-%m-%d")
    Enddate = Begindate + timedelta(days=days[i])
    Enddate = Enddate.strftime("%d/%m/%Y")
    dates.append(Enddate)
```

```python
#list of names of states
statenametupple = []
for s in state:
    k = list(vaccination_data["State_Code"]).index(s)
    v = vaccination_data["State"][k]
    statenametupple.append((s,v))

statename = []    
for i in range(len(statenametupple)):
    statename.append(statenametupple[i][1])
```

```python
#resulted dataframe creation
result = {
    "stateid" : statename,
    "populationleft" : population_left_state,
    "rateofvaccination" : rate_of_vaccination_state,
    "date" : dates
}

result = pd.DataFrame(result)
```
