```python
import pandas as pd
import numpy as np
import json
from datetime import datetime
from time import time
%matplotlib inline 
import matplotlib.pyplot as plt
```

# 1. For district 

### Vaccination count of districts per weeks

```python
#importing vaccination data
vaccine_data = pd.read_csv("data/cowin_vaccine_data_districtwise.csv")
```

```python
vaccine_data
```

```python
#reading neighbors json file as disctionary
with open('data/neighbors-districts-modified.json') as json_file:
    neighbor_data = json.load(json_file)

#list of sorted district in neighbor file  
neighbor_districts = list(neighbor_data.keys())
neighbor_districts = sorted(neighbor_districts)
```

```python
#creating list of weeks
date_list = pd.date_range(start = "1-10-2021",end = "14-08-2021").to_pydatetime().tolist()
for i in range(len(date_list)):
    date_list[i] = date_list[i].strftime("%d/%m/%Y")
    
#week_list contain last date of a week
week_list = []
for i in range(6,len(date_list)+1,7):
    week_list.append(date_list[i])
```

```python
#replacing nan values from dose data
for week in week_list:
    vaccine_data[week + '.3'] = vaccine_data[week + '.3'].fillna(0)
    vaccine_data[week + '.4'] = vaccine_data[week + '.4'].fillna(0)
```

```python
#extracting amount of vaccination of dose1 and dose2 for all weeks 
dose1 = {}
dose2 = {}
for district in neighbor_districts:
    l1 = []
    l2 = []
    k = list(vaccine_data["District_Key"]).index(district)
    l1.append(int(vaccine_data[week_list[0] + ".3"][k]))
    l2.append(int(vaccine_data[week_list[0] + ".4"][k]))
    for i in range(1,len(week_list)):
        l1.append(int(vaccine_data[week_list[i] + ".3"][k]) - int(vaccine_data[week_list[i-1] + ".3"][k]))
        l2.append(int(vaccine_data[week_list[i] + ".4"][k]) - int(vaccine_data[week_list[i-1] + ".4"][k]))
    dose1[district] = l1
    dose2[district] = l2
```

```python
#converting the results into list so that we can crate a dataframe
districtlist = []
timelist = []
dose1list = []
dose2list = []
for district in neighbor_districts:
    for i in range(len(week_list)):
        districtlist.append(district)
        timelist.append(i+1)
        dose1list.append(dose1[district][i])
        dose2list.append(dose2[district][i])
```

```python
#creating required dataframe 
district_result_week = pd.DataFrame(columns = ["districtid","timeid","dose1","dose2"])
district_result_week["districtid"] = districtlist
district_result_week["timeid"] = timelist
district_result_week["dose1"] = dose1list
district_result_week["dose2"] = dose2list
```

```python
district_result_week
```

```python
#exporting the result
district_result_week.to_csv("district-vaccinated-count-week.csv",index = False)
```

# For state 

### Vaccination count of state per weeks

```python
district_result_week
```

```python
#list of states
state = []
for i in neighbor_districts:
    state.append(i.split("_")[0])
state = list(np.unique(state))
```

```python
#adding column for state code
state_code = []
for district in district_result_week["districtid"]:
    state_code.append(district.split("_")[0]) 
    
district_result_week["stateid"] = state_code
```

```python
#creating a dataframe for state wise vaccination of dose1 and dose2 week wise
result = pd.DataFrame(columns = ["stateid","timeid","dose1","dose2"])
for j in range(1,len(week_list)+1):
    for i in state:
        dic = {
            "stateid" : i,
            "timeid" : j,
            "dose1" : np.sum(district_result_week.query("timeid == @j & stateid ==@i")["dose1"]),
            "dose2" : np.sum(district_result_week.query("timeid == @j & stateid ==@i")["dose2"])
        }
        result = result.append(dic,ignore_index=True)

#resetting the index 
result = result.sort_values(["stateid","timeid"])
result.reset_index(drop = True,inplace = True)
```

```python
result
```

```python
result.to_csv("state-vaccinated-count-week.csv",index = False)
```
