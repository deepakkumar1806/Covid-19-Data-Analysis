```python
import pandas as pd
import json
```

```python
#reading neighbors json file as disctionary
with open('data/neighbors-districts-modified.json') as json_file:
    neighbor_data = json.load(json_file)
```

# Preparing a list of district and its neighbours 

```python
ans = []
for k in neighbor_data.keys():
    for v in neighbor_data[k]:
        ans.append((k,v))
    
```

```python
ans
```

```python
df = pd.DataFrame(ans,columns=["District","Neighbour"])
```

```python
df
```

```python
df.to_csv("edge-graph.csv",index=False)
```
