```python
#ratio of female vaccinated to male till 14-08-2021 district wise
female_vaccinated = []
male_vaccinated = []
for district in neighbor_districts:
    k = list(vaccination_data["District"]).index(district[1])
    female_vaccinated.append(int(vaccination_data["14/08/2021.6"][k]))
    male_vaccinated.append(int(vaccination_data["14/08/2021.5"][k]))
    
vaccinationratio = []
for i in range(len(female_vaccinated)):
    vaccinationratio.append(round(female_vaccinated[i]/male_vaccinated[i],2))
```

```python
#Ratio of female population to male district wise
female_population = []
male_population = []
for district in neighbor_districts:
    k = list(census_data["Name"]).index(district[1].lower())
    female_population.append(int(census_data["TOT_F"][k]))
    male_population.append(int(census_data["TOT_M"][k]))
    
populationratio = []
for i in range(len(female_population)):
    populationratio.append(round(female_population[i]/male_population[i],2))
```

```python
#ratio of vaccination ratio to population ratio
ratioofratios = []
for i in range(len(vaccinationratio)):
    ratioofratios.append(round(vaccinationratio[i]/populationratio[i],2))
```

```python
#list of districts according to district ids
districtlist = []
for i in range(len(neighbor_districts)):
    districtlist.append(neighbor_districts[i][0])
```

```python
#resulted dataframe creation
result = {
    "districtid" : districtlist,
    "vaccinationratio" : vaccinationratio,
    "populationratio" : populationratio,
    "ratioofratios" : ratioofratios
}

result = pd.DataFrame(result)
result = result.sort_values("ratioofratios")
```

## Ratio of ratios state wise 

```python
#list of states
state = []
for i in neighbor_districts:
    state.append(i[0].split("_")[0])
state = list(np.unique(state))
```

```python
#Dataframe that contain district wise info
df = {
    "districtid" : districtlist,
    "female_population" : female_population,
    "male_population" : male_population,
    "female_vaccination" : female_vaccinated,
    "male_vaccination" : male_vaccinated
}
df = pd.DataFrame(df)
```

```python
#adding column for state code
state_code = []
for district in df["districtid"]:
    state_code.append(district.split("_")[0]) 
    
df["stateid"] = state_code
```

```python
#state wise information
state_female_population = []
state_male_population = []
state_female_vaccinated = []
state_male_vaccinated = []
for s in state:
    x = df.query("stateid == @s")
    state_female_population.append(np.sum(x["female_population"]))
    state_male_population.append(np.sum(x["male_population"]))
    state_female_vaccinated.append(np.sum(x["female_vaccination"]))
    state_male_vaccinated.append(np.sum(x["male_vaccination"]))
```

```python
#state wise different ratio
state_populationratio = []
state_vaccinationratio = []
state_ratioofratios = []
for i in range(len(state_female_population)):
    state_populationratio.append(round(state_female_population[i]/state_male_population[i],2))
    state_vaccinationratio.append(round(state_female_vaccinated[i]/state_male_vaccinated[i],2))
    state_ratioofratios.append(round(state_vaccinationratio[i]/state_populationratio[i],2))
```

```python
#list of names of states
statenametupple = []
for s in state:
    k = list(vaccination_data["State_Code"]).index(s)
    v = vaccination_data["State"][k]
    statenametupple.append((s,v))

statename = []    
for i in range(len(statenametupple)):
    statename.append(statenametupple[i][1])
```

```python
#resulted dataframe creation
result = {
    "stateid" : statename,
    "vaccinationratio" : state_vaccinationratio,
    "populationratio" : state_populationratio,
    "ratioofratios" : state_ratioofratios
}

result = pd.DataFrame(result)
result = result.sort_values("ratioofratios")
```

```python
result
```

```python
result.to_csv("state-vaccination-population-ratio.csv",index = False)
```

## Ratio of ratios overall 

```python
#all india vaccination and population information
female_population_overall = np.sum(df["female_population"])
male_population_overall = np.sum(df["male_population"])
female_vaccinated_overall = np.sum(df["female_vaccination"])
male_vaccinated_overall = np.sum(df["male_vaccination"])
```

```python
#ratios
populationratio_overall = round(female_population_overall/male_population_overall,2)
vaccinationratio_overall = round(female_vaccinated_overall/male_vaccinated_overall,2)
```

```python
#output data frame
result = pd.DataFrame(columns = ["Overall","vaccinationratio","populationratio","ratioofratios"])
dic = {
    "Overall" : "overall",
    "vaccinationratio" : vaccinationratio_overall,
    "populationratio" : populationratio_overall,
    "ratioofratios" : round(vaccinationratio_overall/populationratio_overall,2)
}
result = result.append(dic,ignore_index= True)
```

```python
result.to_csv("overall-vaccination-population-ratio.csv",index=False)
```
